```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi untuk EDA
def perform_eda(df):
    # Descriptive Statistics
    print("Descriptive Statistics:")
    print(df.describe())
    print()

    # Missing Values
    print("Missing Values:")
    print(df.isnull().sum())
    print()

    # Correlation Heatmap
    print("Correlation Heatmap:")
    corr = df.corr()
    sns.heatmap(corr, annot=True, cmap="coolwarm")
    plt.show()

    # Distribution of Numerical Variables
    print("Distribution of Numerical Variables:")
    numerical_columns = df.select_dtypes(include=np.number).columns
    for column in numerical_columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(df[column], kde=True)
        plt.title(f"Distribution of {column}")
        plt.xlabel(column)
        plt.ylabel("Count")
        plt.show()

    # Countplot of Categorical Variables
    print("Countplot of Categorical Variables:")
    categorical_columns = df.select_dtypes(include='object').columns
    for column in categorical_columns:
        plt.figure(figsize=(10, 4))
        sns.countplot(x=column, data=df)
        plt.title(f"Countplot of {column}")
        plt.xlabel(column)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.show()

# Contoh penggunaan fungsi EDA
df = pd.read_excel('Iris.xlsx')  # Ganti 'data.csv' dengan lokasi dan nama file dataset Anda
perform_eda(df)
```
